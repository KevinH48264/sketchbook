```python
# https://leetcode.com/problems/asteroid-collision/

from typing import List

class Solution:
    def asteroidCollision(self, asteroids: List[int]) -> List[int]:
        """
        Process:
            stack = []
            for a in asteroids:
                a_explode = False # flag

                # check if we need to pop if there's a stack
                while stack:
                    # case 1: a is moving right, we just append
                    if a > 0:
                        break

                    # case 2: a is moving left, we might need to pop
                    elif a < 0:
                        last = stack[-1]

                        if last < 0:
                            stack.append(a)
                        
                        else: # last is moving right
                            if last > a:
                                a_explode = True
                                break # this left moving asteroid will explode
                            else: # a > left
                                stack.pop()

                # add the asteroid
                if not a_explode:
                    stack.append(a)

            return stack

        Result:
        - Finished in 14 mins successfully
        """

        # attempt 2: aiming for a O(1) memory. just use the same array 
        j = 0
        n = len(asteroids)

        for i in range(n):
            a = asteroids[i]

            # handle collisions of j - 1 and a # only if j - 1 >= 0, the j - 1th asteroid is moving right, current a is moving left, and j - 1th asteroid is smaller than the current 
            while j > 0 and asteroids[j - 1] > 0 and a < 0 and asteroids[j - 1] < abs(a):
                j -= 1

            # if j - 1 doesn't exist or our asteroid a is moving right or the j - 1th asteroid is moving left, just set jth to a and increment j
            if j == 0 or a > 0 or asteroids[j - 1] < 0:
                asteroids[j] = a
                j += 1
            elif asteroids[j - 1] == abs(a): # if j-1th asteroid and a collide, then decrement j
                j -= 1
        
        return asteroids[:j]


        # attempt 1
        # stack = []
        # for a in asteroids:
        #     a_explode = False # flag

        #     # check if we need to pop if there's a stack
        #     while stack and a < 0 and stack[-1] > 0: # we only collide if there's asteroids in the past, the current one is moving left, and the latest value is moving right, otherwise there will be no collisions
        #         if abs(stack[-1]) > abs(a): # magnitude of last is greater than magnitude of a
        #             a_explode = True
        #             break # this left moving asteroid will explode
        #         elif abs(stack[-1]) == abs(a): # both will explode # magnitudes are the same
        #             a_explode = True
        #             stack.pop()
        #             break
        #         else: # a > last
        #             stack.pop()

        #     # add the asteroid
        #     if not a_explode:
        #         stack.append(a)
                
        # return stack

        # # Time: O(n)
        # # Space: O(n)
```
